##### Пропуски в столбцах `floors_total`, `locality_name`, `balcony`
Выведем информацию о пропущенных значениях для каждого столбца, используя комбинацию методов `isna()` и `sum()`.

В связи с небольшим количеством пропущенных значений в этих столбцах можно удалить строки с пропусками.

```python
#Для замены пропущенных значений применим метод `fillna()`
data['balcony'] = data['balcony'].fillna(value=0)

#Проверим наличие пропусков в столбце `balcony`
data['balcony'].isna().sum()
```

##### Пропуски в столбцах `living_area`, `kitchen_area`
Пропуски в столбцах kitchen_area и living_area могли появиться из-за того, что продавец указал только общую площадь квартиры, а информации о площади кухни и жилой площади у него не оказалось в этот момент. Жилая площадь и площадь кухни квартир больше всего зависит от общей площади квартиры. Можно сгруппировать квартиры по размеру общей площади и заменить пропущенные значения в столбцах kitchen_area и living_area медианными значениями по группам.

```python
#Выполним категоризацию данных по общей площади квартир
#Для этого создадим таблицу `data_area`, выполним срез из `data` по трём нужным столбцам
data_area = data.loc[:, ['total_area', 'living_area', 'kitchen_area']]

#Выведем первые строки новой таблицы
print(data_area.head())
```

```python
#Посмотрим, какие в столбце `total_area` минимальное и максимальное значения
print('Минимальное значение общей площади', data_area['total_area'].min(),
      '\nМаксимальное значение общей площади', data_area['total_area'].max())
```

Создадим в новой таблице `data_area` столбец `area_category` со следующими категориями общей площади:
* до 20 - `E`
* 20-50 - `D`
* 50-100 - `C`
* 100-200 - `B`
* более 200 - `A`
Для этого применим собственную функцию `cat_area` и метод `apply()`

```python
# Создаём функцию для категоризации
def cat_area(area):
    try:
        if area <= 20:
            return 'E'
        if area <= 50:
            return 'D'
        if area <= 100:
            return 'C'
        if area <= 200:
            return 'B'
        return 'A'
    except:
        pass
```

```python
# Применяем функцию к столбцу
data_area['area_category'] = data_area['total_area'].apply(cat_area)
print(data_area.head())
```

```python
#Выполним группировку в таблице `data_area` по категориям площади.
#Для каждой группы посчитаем медианное значение общей площади, жилой площади и площади кухни.
#Для группировки применяем методы `groupby()` и agg()
data_area_grouped = data_area.groupby('area_category').agg({'total_area': 'median', 'living_area': 'median', 'kitchen_area': 'median'})
data_area_grouped
```

```python
#Вычислим в каждой группе среднюю долю жилой площади в общей площади (столбец `part_living`)
#и среднюю долю площади кухни в общей площади (столбец `part_kitchen`)
data_area_grouped['part_living'] = data_area_grouped['living_area'] / data_area_grouped['total_area']
data_area_grouped['part_kitchen'] = data_area_grouped['kitchen_area'] / data_area_grouped['total_area']
data_area_grouped.head()
```

Теперь мы знаем по каждой категории, какую долю в квартире в среднем занимает жилая площадь и какую - кухня. Занесем эти данные в таблицу data_area, распространяя полученные значения на каждую квартиру.

```python
#Методом merge() присоединим к датафрейму `data_area` два столбца из датафрейма `data_area_grouped` (по столбцу `area_category`)
data_area = data_area.merge(data_area_grouped.loc[:, ['part_living', 'part_kitchen']] , on='area_category', how='left')
data_area.head()
```

Теперь мы можем заполнить пропуски в столбцах `living_area` и `kitchen_area`.

```python
#Методом fillna() находим каждый пропуск в столбце `living_area`, заменяем его значением, полученным путем умножения общей площади
#на долю жилой площади в общей площади, округляя это значение до двух знаков после точки
data_area['living_area'] = data_area['living_area'].fillna(value=data_area['total_area'] * data_area['part_living']).round(2)

#Аналогично заменяем пропуски в столбце `kitchen_area`
data_area['kitchen_area'] = data_area['kitchen_area'].fillna(value=data_area['total_area'] * data_area['part_kitchen']).round(2)
print(data_area['kitchen_area'].isna().sum(), data_area['kitchen_area'].isna().sum())
```

```python
data_area.head()
```

Заменим в датафрейме `data` столбцы `living_area` и `kitchen_area` аналогичными столбцами из датафрейма `data_area`. Так мы избавимся от пропусков в этих столбцах, заменив их значениями, рассчитанными при помощи категоризации данных.

```python
data['living_area'] = data_area['living_area']
data['kitchen_area'] = data_area['kitchen_area']
data.head(20)
```

```python
#Проверим ещё раз, есть ли пропущенные значения в столбцах `living_area` и `kitchen_area`
print(data['living_area'].isna().sum(), data['kitchen_area'].isna().sum())
```

Проверим, не превышает ли сумма жилой площади и площади кухни общую площадь квартиры. Сделаем срез методом `query()` объектов, в которых имеет место такая ошибка.

```python
mistake = data.query('(living_area + kitchen_area) >= total_area')
mistake
```

Действительно, 194 объекта получили заполнение пропущенных значений некорректно. Придётся их тоже удалить.

```python
data = data.query('(living_area + kitchen_area) < total_area')

#Проверим объем оставшихся в таблице данных
print(len(data))
```

##### Пропуски в столбце `ceiling_height`

В столбце `ceiling_height` пропуски могут быть также из-за того, что у продавца под рукой не оказалось информации о высоте потолков. Эти пропущенные значения также можно заменить медианным значением по данному столбцу.

### Изменение типов данных

Судя по общей информации о датафрейме, нужно обратить внимание на типы данных в столбцах `balcony`, `floors_total` и `first_day_exposition`.
Столбец `balcony` в исходных данных имеет тип `float`. Но число балконов не может быть дробным, оно может быть только целочисленным. Изменим тип данных на `int` с помощью метода `astype()`. Проверим результат замены, вызвав метод `value_counts()`.

```python
data['balcony'] = data['balcony'].astype('int')
data['balcony'].value_counts()
```

Столбец `floors_total` также имеет тип `float`, однако число этажей не может быть дробным, оно может быть только целочисленным. Изменим тип данных на `int` с помощью метода `astype()`. Проверим результат замены, выведя первые пять строк таблицы.

```python
data['floors_total'] = data['floors_total'].astype('int')
data.head()
```

В столбце `days_exposition` также неподходящий тип данных. Нам для исследования лучше иметь целочисленные значения количества дней размещения объявления от публикации до снятия. Но пропущенные значения, которые мы не стали восстанавливать, не позволяют это сделать по отношению ко всему столбцу. Можно использовать особый nullable тип данных, который поддерживает пропуски:
`.astype('Int64')`.

```python
data['days_exposition'] = data['days_exposition'].round().astype('Int64')
data.head()
```

Данные в столбце `first_day_exposition` имеют тип `object`. Их нужно перевести в тип `datetime`, чтобы можно было далее работать с датами.

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
data.head()
```

## Добавление в таблицу новых столбцов

### Цена одного квадратного метра

Добавим в таблицу столбец `price_one_square` - цена одного квадратного метра. Для этого поделим стоимость объекта на его общую площадь, а затем округлим до двух знаков после запятой.

```python
data['price_one_square'] = (data['last_price'] / data['total_area']).round(2)
data.head()
```

### День недели, месяц и год публикации

Добавим в таблицу столбцы `weekday_exp` - день недели публикации объявления, `month_exp` - месяц публикации объявления и `year_exp` - год публикации объявления.

```python
data['weekday_exp'] = data['first_day_exposition'].dt.weekday
data['month_exp'] = data['first_day_exposition'].dt.month
data['year_exp'] = data['first_day_exposition'].dt.year
data.head()
```

### Тип этажа квартиры

Добавим в таблицу столбец `floor_group` с типом этажа квартиры: `первый`, `последний`, `другой`. Для этого применим метод `where()`.

```python
#Сначала всем строкам нового столбца присваиваем категорию 'другие'
data['floor_group'] = 'другие'

#Затем для первых этажей заменим категорию на 'первый'
data['floor_group'] = data['floor_group'].where(data['floor'] != 1, 'первый')

#И наконец для последних этажей заменим категорию на 'последний'
data['floor_group'] = data['floor_group'].where(data['floor'] != data['floors_total'], 'последний')

data.head(20)
```

```python
data['centers_nearest_km'] = (data['cityCenters_nearest'] / 1000).round()
data['centers_nearest_km'] = data['centers_nearest_km'].astype('Int64')
data.head()
```

* есть все нужные столбцы,
* в важных столбцах нет пропущенных значений,
* индексы идут по порядку,

Построим гистограмму по столбцу `ceiling_height` таблицы `good_data`.

```python
good_data['ceiling_height'].hist(bins=30)
plt.title('Распределение квартир по высоте потолков')
plt.xlabel('Высота потолков')
plt.ylabel('Количество квартир')
plt.show()
```

```python
print(good_data['ceiling_height'].sort_values(ascending=False).head(20))
```

```python
print(good_data['ceiling_height'].sort_values(ascending=False).tail(20))
```

Действительно, высота потолка 100 метров - это слишком большое значение. И все следующие по убыванию значения тоже слишком велики. Высота потолка обычно не превышает 3,5 метров. Минимальная высота потолка в России, согласно строительным нормам, 2,5 метра. Остальные значения неправдоподобны. Можно предположить, что значения от 25,0 до 32,0 могли появиться вследствие ошибки пользователя. Он мог просто не поставить запятую, и вместо вещественного числа получилось целое. Обработаем эти значения, уменьшив их в 10 раз. Остальным аномальным значениям (больше 32,0 и меньше 2,5 метров, а также от 3,5 до 25,0 метров логической замены предложить невозможно. ### Стоимость квадратного метра в отдельных населённых пунктах

##### Населенные пункты с наибольшим числом объявлений
Определим, в каких населённых пунктах было размещено наибольшее количество объявлений о продаже квартир. Для этого выполним группировку данных в таблице `good_data` по столбцу `locality_name` методом `groupby()`, посчитаем число объявлений в каждом населенном пункте методом `count()` и отсортируем полученный `Series` в порядке убывания значений методом `sort_values()`.

```python
name_group = good_data.groupby('locality_name')['locality_name'].count().sort_values(ascending=False)

#Выведем на экран 10 населенных пунктов с максимальным числом объявлений
name_group.head(10)
```

```python
#При помощи логической индексации оставим только 10 строк с самыми большими значениями числа объявлений
name_group_10 = name_group.loc[:'Выборг']
name_group_10
```

```python
#Выбираем строки датафрейма `good_data`, содержащие в столбце `locality_name` значения из индексов `Series` `name_group_10`
good_data_locality = good_data.query(
    'locality_name in @name_group_10.index'
)
#Проверим, выведя уникальные значения в столбце `locality_name` получившегося датафрейма `good_data_locality`
good_data_locality['locality_name'].unique()
```

##### Стоимость квадратного метра в выбранных населенных пунктах
Составим сводную таблицу для выбранных населенных пунктов с наибольшим числом обявлений, где для этих населенных пунктов выведем медианную стоимость одного квадратного метра.

```python
#Применяем метод `pivot_table()
locality_name_group = good_data_locality.pivot_table(
    index='locality_name',
    values='price_one_square',
    aggfunc=['count', 'median']
)
locality_name_group
```

```python
#Переименуем столбцы
locality_name_group.columns = ['count', 'price_one_square']

#Выполним сортировку населенных пунктов в порядке убывания цены одного квадратного метра
locality_name_group = locality_name_group.sort_values(by='price_one_square', ascending=False)
locality_name_group
```

**Вывод.** Среди 10 населенных пунктов с наибольшим количеством объявлений самая высокая цена одного квадратного метра жилья в Санкт-Петербурге - более 103 тысяч. Самая низкая стоимость жилья среди выбранных населенных пунктов - в Выборге - 58 тысяч.

```python
data_spb = good_data.query('locality_name == "Санкт-Петербург"')

#Проверим, вызвав перечень уникальных значений столбца `locality_name`
data_spb['locality_name'].unique()
```

```python
#Проверим, какие есть уникальные значения в столбце `centers_nearest_km` таблицы `data_spb`
data_spb['centers_nearest_km'].unique()
```

Самая низкая цена на квартиры на расстоянии 26 километров от центра Санкт-Петербурга. Далее на расстоянии 27-28-29 километров отмечен рост цен на жильё. Возможно, это какие-то новые комплексы дорогих коттеджей. Однако общая тенденция уменьшения цены квартир с увеличением расстояния от центра явно прослеживается.

Далее были изменены типы данных в некоторых столбцах для облегчения дальнейшей работы с ними. Кроме того, были удалены дубликаты путем приведения к однотипным видам.

Далее мы добавили в таблицу дополнительные столбцы с расчетными данными, которые нам нужны для исследования. Кроме того, данные были исследованы более подробно и удалены или исправлены редкие и аномальные значения.

После проведения предобработки данных из 23699 строк в таблице осталось 22462 строки, то есть объем данных изменился незначительно, но эти данные стали пригодны для исследования.

Изучение времени продажи квартир показало, что 50 % квартир было продано в срок от 1,5 до 7 месяцев. 25 % квартир было продано в первые 40-45 дней после размещения объявления. По всей видимости, цена на эти квартиры была занижена. Долгими продажами квартир можно считать срок продажи от 230 до 500 дней. Если квартира продавалась более 500 дней - это аномалия. Для продавцов этих квартир это повод скорректировать объявление, и, прежде всего, цену квартиры. 

Исследование показало прямую зависимость цены квартир от площади (общей, жилой и кухни), от количества комнат. Большую роль в стоимости квартиры играет этаж: на первом и последнем этажах квартиры стоят дешевле, чем на других этажах.

День недели размещения объявления роли в цене не играет, а вот месяц размещения важен: так, квартиры, которые выставлены на продажу в апреле, в среднем стоят дороже, чем квартиры, выставленные в июне. По годам также есть разница: в рассматриваемом периоде самыми высокими цены на квартиры были в 2014 году, далее цены снижались вплоть до 2017 года, и только в 2019 году опять наблюдался рост цен на жильё.

Среди населенных пунктов с наибольшим количеством объявлений самая высокая цена одного квадратного метра жилья в Санкт-Петербурге - более 103 тысяч. Второе место по стоимости жилья занимает Пушкин, третье - поселок Парголово. В непосредственной близости к центру Санкт-Петербурга (в 1 километре от центра) самая высокая средняя цена на квартиры (более 9 миллионов). По мере удаления от центра Санкт-Петербурга наблюдается тенденция уменьшения стоимости жилья.

Итак, нами проведён исследовательский анализ стоимости объектов недвижимости, который позволит построить автоматизированную систему, способную отследить аномалии и мошенническую деятельность.
